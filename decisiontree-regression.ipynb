```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt

# Define the dataset
X = np.array([[2], [3], [4], [5], [6]])
y = np.array([3, 4, 2, 5, 7])

# Reshape X if necessary (sklearn expects a 2D array for features)
X = X.reshape(-1, 1)

# Create the DecisionTreeRegressor model
model = DecisionTreeRegressor(max_depth=1)

# Train the model on the dataset
model.fit(X, y)

# Make predictions on the training data
predictions = model.predict(X)

# Print the predictions
print("Predictions:", predictions)

# Generate a sequence of numbers for plotting the decision boundary
X_plot = np.arange(2, 7, 0.01).reshape(-1, 1)
y_plot = model.predict(X_plot)

# Plot the actual data points
plt.scatter(X, y, color='red', label='Actual data')

# Plot the model's predictions
plt.plot(X_plot, y_plot, color='blue', label='Model prediction')

# Add labels and legend
plt.xlabel('Feature (X)')
plt.ylabel('Target (Y)')
plt.title('Decision Tree Regression')
plt.legend()

# Show the plot
plt.show()
```
