```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate sample data
np.random.seed(0)
X = 2 - 3 * np.random.normal(0, 1, 100)
y = 4 + 2 * X + 3 * X**2 + np.random.normal(0, 2, 100)

# Scatter plot of the data
plt.scatter(X, y, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Sample Data')
plt.show()

# Transform the data to include polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X.reshape(-1, 1))

# Fit the linear regression model
model = LinearRegression()
model.fit(X_poly, y)

# Print the model parameters
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")

# Make predictions
y_pred = model.predict(X_poly)

# Plot the results
plt.scatter(X, y, color='blue', label='Data')
plt.plot(np.sort(X), y_pred[np.argsort(X)], color='red', label='Polynomial Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression Fit')
plt.legend()
plt.show()

# Compute and print the mean squared error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
X = 2 - 3 * np.random.normal(0, 1, 100)
y = 4 + 2 * X + 3 * X**2 + np.random.normal(0, 2, 100)

# Scatter plot of the data
plt.scatter(X, y, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Sample Data')
plt.show()

# Transform the data to include polynomial features
degree = 1
X_poly = np.vstack([X**i for i in range(degree + 1)]).T

# Compute the model parameters using the normal equation
X_poly_transpose = X_poly.T
theta = np.linalg.inv(X_poly_transpose.dot(X_poly)).dot(X_poly_transpose).dot(y)

# Print the model parameters
print(f"Model parameters (theta): {theta}")

# Make predictions
y_pred = X_poly.dot(theta)

# Plot the results
plt.scatter(X, y, color='blue', label='Data')
plt.plot(np.sort(X), y_pred[np.argsort(X)], color='red', label='Polynomial Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression Fit')
plt.legend()
plt.show()

# Compute and print the mean squared error
mse = np.mean((y - y_pred)**2)
print(f"Mean Squared Error: {mse}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
x1 = np.random.normal(0, 1, 100)
x2 = np.random.normal(0, 1, 100)
y = 3 + 2*x1 + 4*x2 + 5*x1**2 + 6*x1*x2 + 7*x2**2 + np.random.normal(0, 1, 100)

# Scatter plot of the data
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, y, color='blue')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
plt.title('Sample Data')
plt.show()

# Transform the data to include polynomial features up to degree 2
degree = 2
def polynomial_features(x1, x2, degree):
    features = [np.ones(len(x1))]  # Bias term (intercept)
    for i in range(1, degree + 1):
        for j in range(i + 1):
            features.append((x1 ** (i - j)) * (x2 ** j))
    return np.vstack(features).T

X_poly = polynomial_features(x1, x2, degree)

# Compute the model parameters using the normal equation
X_poly_transpose = X_poly.T
theta = np.linalg.inv(X_poly_transpose.dot(X_poly)).dot(X_poly_transpose).dot(y)

# Print the model parameters
print(f"Model parameters (theta): {theta}")

# Make predictions
y_pred = X_poly.dot(theta)

# Scatter plot of the data and the polynomial fit
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, y, color='blue', label='Data')
ax.scatter(x1, x2, y_pred, color='red', label='Polynomial Fit')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
plt.title('Polynomial Regression Fit')
plt.legend()
plt.show()

# Compute and print the mean squared error
mse = np.mean((y - y_pred)**2)
print(f"Mean Squared Error: {mse}")
```
